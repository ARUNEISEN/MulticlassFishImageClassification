```python
from DataAugmentation import data_loader
import os
root_data_dir = r"D:\Projects\MulticlassFishImageClassification\Project_data\data"

train_dir = os.path.join(root_data_dir, 'train')
valid_dir = os.path.join(root_data_dir, 'val')
test_dir = os.path.join(root_data_dir, 'test')

train_generator, valid_generator, test_generator = data_loader(train_dir,valid_dir,test_dir)
```

### MODEL Traing Pipeline


```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.applications import (
    VGG16, ResNet50, MobileNetV2, InceptionV3, EfficientNetB0
)

NUM_CLASSES = train_generator.num_classes
IMG_SIZE = (256, 256)
EPOCHS = 15

callbacks = [
    EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True),
    ModelCheckpoint("best_model.h5", monitor="val_accuracy", save_best_only=True)
]
```

### CNN Model From Scratch

```python
def build_cnn(num_classes):
    model = tf.keras.Sequential([
        layers.Input(shape=(256,256,3)),
        
        layers.Conv2D(32, 3, activation="relu"),
        layers.MaxPooling2D(),

        layers.Conv2D(64, 3, activation="relu"),
        layers.MaxPooling2D(),

        layers.Conv2D(128, 3, activation="relu"),
        layers.MaxPooling2D(),

        layers.Flatten(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax")
    ])

    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model
```

```python
from tensorflow.keras.models import Model
```

### VGG16 Model From Scratch

```python
def build_vgg16(num_classes):
    base = VGG16(include_top=False, weights="imagenet", input_shape=(256,256,3))
    base.trainable = False

    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)  # new layer instance
    x = layers.Dropout(0.5)(x)
    
    output = layers.Dense(num_classes, activation="softmax")(x)

    model = Model(inputs=base.input, outputs=output)
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model
```

### ResNet50 Model From Scratch

```python
def build_resnet50(num_classes):
    base = ResNet50(include_top=False, weights="imagenet", input_shape=(256,256,3))
    base.trainable = False

    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)

    output = layers.Dense(num_classes, activation="softmax")(x)

    model = Model(inputs=base.input, outputs=output)
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model
```

### MobileNetV2 Model From Scratch

```python
def build_mobilenet(num_classes):
    base = MobileNetV2(include_top=False, weights="imagenet", input_shape=(256,256,3))
    base.trainable = False

    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)

    output = layers.Dense(num_classes, activation="softmax")(x)

    model = Model(inputs=base.input, outputs=output)
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model
```

### InceptionV3 Model From Scratch

```python
def build_inceptionv3(num_classes):
    base = InceptionV3(include_top=False, weights="imagenet", input_shape=(256,256,3))
    base.trainable = False

    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)

    output = layers.Dense(num_classes, activation="softmax")(x)

    model = Model(inputs=base.input, outputs=output)
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model
```

### Train CNN Model

### Evaluate all trained Model

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

def evaluate_model(model, name):
    print(f"Evaluating {name}...")

    loss, acc = model.evaluate(test_generator)
    preds = model.predict(test_generator)
    y_pred = np.argmax(preds, axis=1)
    y_true = test_generator.classes

    print("\nAccuracy:", acc)
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred))

    return acc
```

```python
results = {}

for model_name in ["CNN", "VGG16", "ResNet50", "MobileNetV2", "InceptionV300"]:
    model = tf.keras.models.load_model(f"D:/Projects/MulticlassFishImageClassification/Models/{model_name}_model.h5")
    acc = evaluate_model(model, model_name)
    results[model_name] = acc

best_model = max(results, key=results.get)
best_model, results[best_model]
```
